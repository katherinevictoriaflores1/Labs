```python
# 13.4
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.read_csv('/content/SQC_HW4_data.csv')

rep1 = df.copy()
rep1['Y'] = df['Response']
rep2 = df.copy()
rep2['Y'] = df['Response2']
df_long = pd.concat([rep1, rep2], ignore_index=True)

model = ols('Y ~ A + B + C + A:B + A:C + B:C + A:B:C', data=df_long).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
print("\n ANOVA Table ")
print(aov_table)

print(model.summary())
```

```python
#13.6

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


df = pd.read_csv('/content/SQC_HW_data2.csv')

rep1 = df.copy()
rep1['Y'] = df['Response1']

rep2 = df.copy()
rep2['Y'] = df['Response2']

df_long = pd.concat([rep1, rep2], ignore_index=True)


model = ols('Y ~ A + B + C + D + A:B + A:C + A:D + B:C + B:D + C:D + \
            A:B:C + A:B:D + A:C:D + B:C:D + A:B:C:D', data=df_long).fit()


aov_table = sm.stats.anova_lm(model, typ=2)
print("\n===== ANOVA Table =====")
print(aov_table)


print("\n===== Model Summary =====")
print(model.summary())
```
