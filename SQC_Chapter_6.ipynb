```python
## Katherine Flores Chapter 6 HW
## 6.8, 6.25, 6.63, and 6.64


##6.8

import statistics
import matplotlib.pyplot as plt
import numpy as np

# Data setup
phase_1_dataset = [
    [6, 9, 10, 15],
    [10, 4, 6, 11],
    [7, 8, 10, 5],
    [8, 9, 6, 13],
    [9, 10, 7, 13],
    [12, 11, 10, 10],
    [16, 10, 8, 9],
    [7, 5, 10, 4],
    [9, 7, 8, 12],
    [15, 16, 10, 13],
    [8, 12, 14, 16],
    [6, 13, 9, 11],
    [16, 9, 13, 15],
    [7, 13, 10, 12],
    [11, 7, 10, 16],
    [15, 10, 11, 14],
    [9, 8, 12, 10],
    [15, 7, 10, 11],
    [8, 6, 9, 12],
    [13, 14, 11, 15]
]

# Define lists for group means and ranges
x_bar_phase_1 = []
r_phase_1 = []

# Get and append group means and ranges
for group in phase_1_dataset:
    x_bar_phase_1.append(statistics.mean(group))
    r_phase_1.append(max(group) - min(group))

m = len(phase_1_dataset)
print('Groups in phase_1_dataset:', m)

# Chart parameters for n=4
A2 = 0.729
D3 = 0
D4 = 2.282

# Plot x-bar and R charts
fig, axs = plt.subplots(2, figsize=(15, 15))

# X-bar Chart
axs[0].plot(x_bar_phase_1, linestyle='-', marker='o', color='black')
axs[0].axhline(statistics.mean(x_bar_phase_1) + A2 * statistics.mean(r_phase_1), color='red', linestyle='dashed')
axs[0].axhline(statistics.mean(x_bar_phase_1) - A2 * statistics.mean(r_phase_1), color='red', linestyle='dashed')
axs[0].axhline(statistics.mean(x_bar_phase_1), color='blue')
axs[0].set_title('X-bar Chart')
axs[0].set(xlabel='Group', ylabel='Mean')

# R Chart
axs[1].plot(r_phase_1, linestyle='-', marker='o', color='black')
axs[1].axhline(D4 * statistics.mean(r_phase_1), color='red', linestyle='dashed')
axs[1].axhline(D3 * statistics.mean(r_phase_1), color='red', linestyle='dashed')
axs[1].axhline(statistics.mean(r_phase_1), color='blue')
axs[1].set_ylim(bottom=0)
axs[1].set_title('R Chart')
axs[1].set(xlabel='Group', ylabel='Range')

plt.tight_layout()
plt.show()

# Display control limits numerically
Xbar_bar = statistics.mean(x_bar_phase_1)
R_bar = statistics.mean(r_phase_1)
UCLx = Xbar_bar + A2 * R_bar
LCLx = Xbar_bar - A2 * R_bar
UCLR = D4 * R_bar
LCLR = D3 * R_bar

print(f"\nX-bar Chart Limits:")
print(f"UCLx = {UCLx:.3f}, CL = {Xbar_bar:.3f}, LCLx = {LCLx:.3f}")
print(f"R Chart Limits:")
print(f"UCLR = {UCLR:.3f}, CL = {R_bar:.3f}, LCLR = {LCLR:.3f}")

```

1. Is the process in statistical control?)
After plotting the x-bar and R charts, all the points stayed inside the control limits. There werent any weird patterns or runs, so that means the process is in statistical control. The variation we see is just random and not from any special cause.

2. Process capability)
Since the process is stable, we can look at capability. The specs are 350  5 volts, which means the range is from 345 to 355. when converting the sample data back to actual voltage (by dividing by 10 and adding 350), the variation is pretty small compared to the spec range. This suggests the process can meet the voltage requirements, so its capable.

3. Is there evidence to prove this process is in control? (Voltage variability))
Yes there is. Both charts show consistent behavior with no points outside the limits and no clear trends. That supports the assumption that the process has been in control during the time we collected the samples.
